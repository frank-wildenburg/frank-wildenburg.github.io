# Gödel, Escher, Bach

Starting in the summer of 2025, Wouter van Etteger and I co-organized a reading group for the classic book _Gödel, Escher, Bach: an Eternal Golden Braid_ by Douglas Hofstadter. Our intent was to read through the book in eight months, discussing the book through four digital meetings and a group chat, and finishing by visiting a performance of _Die Kunst der Fuge_ by Bach, one of the works discussed in the book. On this page,you can find the chapter notes I wrote for this group.

## Part 1: GEB

<details>
  <summary><i>Introduction: A Musico-Logical Offering</i></summary>

The book starts with an introduction, although this introduction may just as well be seen as an executive summary of the main themes of the book. One of the main parts of the introduction is introducing the main 'subjects' of the book, all of whom will be extensively covered throughout the rest of the book:

* Bach, the musician, known (amongst other things) for creating fugas
* Escher, the artist, known for creating (seemingly) paradoxical artworks
* Gödel, the logician, known for a Gödel's theorem, which says formal systems cannot be both complete and consistent

One of the main characteristics linking the above is the concept of the Strange Loop. Hofstadter defines such a loop as a (hierarchical) system where we can arrive at the starting 'level' by consistently 'heading in one direction'. We can see this in Bach (arriving back at a note after going consistently up from this note), in Escher (returning at a previous point of a staircase after only going up), and in Gödel (by representing statements about numbers as numbers themselves). 

Hofstadter then describes how this concept has been a challenge to (mathematical) logic in the previous century, finally arriving at computers and the concept of (artificial) intelligence. Hofdstadter's main claim here is that Strange Loops form the core of the concept of intelligence.

Finally, Hofstadter describes the structure of the book, which consists of Dialogues (inspired by Zeno and Lewis Carroll) metaphorically introducing a topic, followed by Chapters that cover the topic more in depth and abstractly

Reading tip: when describing Bach, Hofstadter introduces the term 'ricercar', a type of fugue, but also meaning 'searching'. This term refers to riddles and games hidden in Bach's fugas, but just as much to GEB itself. Indeed, the book contains many hidden messages, some of which are later made explicit, while others remain hidden. When reading the book, it is a good exercise to try to spot these riddles. For example, can you find the link to (the name of) Chapter 1 hidden in the Three-Part Intervention?
</details>

<details>
  <summary><i>Chapter I: The MU-puzzle</i></summary>
  
  Chapter 1 introduces the concept of formal systems and ways of thinking about them, using the so-called MIU-system. In essence, a formal system consists of 1) starting theorems, called axioms, and 2) rules for deriving new theorems from existing theorems. A step-by-step demonstration of deriving a theorem is called a derivation. A derivation also proves its final theorem.

  One can reason about formal systems both from inside and outside the system. Outside (and only outside) of a formal system we can use our intelligence to reason about a system - but we do so using knowledge that is not contained in the system! For humans, reasoning 'outside' the system is done almost automatically

  Reading tip: the idea of reasoning inside and outside of a system will be used throughout the book, so try to think about what these terms mean here. For example, which part of a decision procedure is 'inside' a formal system, and which part is 'outside'?

  Finally, Hofstadter introduces the concept of a decision procedure. Such a procedure, when given a theorem, always produces a proof of that theorem in a finite amount of time if one exists.

  Reading tip : the question of this chapter is: can we create "MU" from "MI"? After reading the chapter, you probably know the answer. But suppose you wanted to prove this, how would you do so? What would a formal system need to create such a proof?
</details>

<details>
  <summary><i>Chapter II: Meaning and Form in Mathematics</i></summary>
  
  This chapter discusses the link between formal systems and reality; the 'meaning' of formal systems. Hofstadter does so by introducing a new formal system, the pq-system, that corresponds to some part of reality.

  An important distinction made by Hofstadter is that a statement corresponding to some part of reality does not mean that statement 'means' that part of reality. Rather, we say that there is an isomorphism between a system and reality - some correspondence, or translation, that preserves some structure (in this case, the correspondence that all theorems are true when interpreted).

  The specific correspondence chosen is called an interpretation. An interpretation can be meaningful (corresponding to some part of reality), or meaningless (in which case theorems sound no 'better' or 'true' than non-theorems). 

  We can also distinguish active and passive meanings of a system. If a meaning is active, we can use it to create new rules in the system. If a meaning is passive, we may choose a meaning (e.g., based on isomorphisms), but not create new rules/theorems for the system. For example, in a formal system, we may not create new rules or theorems based on their meaning even if they seem true! Doing so would result in a different formal system, after all.

  Of course, we would like our formal systems to be as meaningful as possible, i.e., correspond to reality as much as possible. For example, as a mathematician, we would like to have a formal system representing ideal numbers to be as close to 'real' numbers as possible. But nevertheless, reality and formal systems are independent from each other - only the interpretation of the formal system through an isomorphism connects them.

  Reading tip: the book (for now) mainly considers formal systems of mathematics, but formal systems exist in all parts of the world. For example, the weather prediction as a formal system of the real weather. Try to think about why you can consider things like this a formal system.
</details>

<details>
  <summary><i>Chapter III: Figure and Ground</i></summary>
  
  Last chapter's system captures concepts using only typographical manipulations (i.e., without needing to consider the meaning of the underlying concept). This chapter shows that such a typographical system can capture more complicated concepts, such as the concept of prime numbers.

  When doing so, it's tempting to use something 'not being a theorem' as a fact in proving something else being a theorem. For example, using something not being divisible as a way to prove something is prime. Yet doing so is an oversimplification, and in some cases, not possible at all.

  Hofstadter illustrates this using a comparison with art. In this comparison, theorems can be seen as the positive space of a work, and non-theorems as the negative space. Yet just as there negative spaces that are not the positive space of some other work, there are also formal systems whose set of non-theorems are not the set of theorems of any formal system (formally: some recursively enumerable sets are not recursive).

  Reading tip: Hofstadter explains _why_ this happens later in the book, so for now, it's fine to just take it as a given fact - but do try to understand what this fact expresses, and why that matters. 

  (This also means, as a logical consequence, that these formal systems do not have a typographical decision procedure).

  Luckily, prime numbers are not such a set, and so the chapter ends by giving a decision procedure in our formal system. Appropriately, this procedure does not depend on some recursive definition (i.e., being divisible, or in other words, being non-prime) but defines non-divisibility directly.

  Reading tip: have you deciphered the contracrostipunctus? And if you have: have you deciphered it?
</details>

<details>
  <summary><i>Chapter IV: Consistency, Completeness, and Geometry</i></summary>
  
  In very broad terms, this chapter is about properties that we might want (an interpretation of) a formal system to have: consistency and completeness. 

  Consistency can either mean that the formal system doesn't contradict the real world (external consistency), or that it doesn't contradict itself (internal consistency).  For example, a formal system might be externally inconsistent if it says that Mount Everest isn't the highest mountain, and internally inconsistent if Mount Everest both is and isn't the highest mountain at the same time. Completeness, meanwhile, says that all real-world statements expressible in our formal system are also theorems of that formal system. 

  Hofstadter emphasizes that both completeness and consistency are properties of the interpretation of a formal system, not the formal system itself. For example, two theorems that seem contradictory (and would therefore break internal consistency) might be perfectly fine when given the right interpretation.

  However, it is difficult to tell whether we have the right interpretation. This is especially difficult when concepts in our formal system are also 'normal' words used in everyday language. Hofstadter illustrates this using non-Euclidean geometry as an example. For many years, people thought these geometries were inconsistent, only to eventually find out that they are not when we consider the right meaning of terms like 'point' and 'line'.

  An important theorem about completeness and consistency is Gödel's Theorem, which says that there is no complete and consistent 'strong' axiomatic system. This theorem is isomorphically illustrated by the Contracrostipunctus. For some, this is seen as a big problem. Hofstadter, however, argues that this simply means we have unrealistic expectations.

  Reading tip: the mapping between the Contracrostipunctus and Gödel's theorem is really the key of this chapter. If you want, you can try rereading the Contracrostipunctus with the given mapping in mind.
</details>

<details>
  <summary><i>Chapter V: Recursive Structures and Processes</i></summary>
  
  Chapter 5 concerns the concept of recursion. Recursion is a property where something (a sentence, a song, etc.) contains a smaller, possibly varied version of itself. 

  Hofstadter starts by introducing some terminology, originating from computer science. When dealing with recursion, 'push' means to 'rise a level deeper' - to start a new task, while remembering the details of your old task. To 'pop' is the opposite: to finish work on one level and 'return' to the point you left on a previous level. Information about these levels - which tasks are involved and where you are in these tasks - is stored on a 'stack'.

  Hofstadter then gives many examples of recursion:
  * In music (like Bach's Kleines harmonisches Labyrinth)
  * In language (like a noun which contains a noun somewhere, like "the strange bagels that *the purple cow without horns* gobbled")
  * In mathematics (like the Fibonacci sequence)
  * In physics (properties of the behaviour of particles)
  * In art (Escher)
  * In games (strategies in Chess)
  These concepts can also be graphically represented, for example with Recursive Transition Networks (RTNs).

  In some way, the 'levels' of recursion are 'the same', in that they share some properties. At the same time, however, they do differ. In particular, a new level of recursion is almost always going to be _simpler_ than a previous level, and it's this that allows recursive behaviours to 'bottom out' and finish - i.e., this is what makes a recursive definition different from a circular definition.
</details>

<details>
  <summary><i>Chapter VI: The Location of Meaning</i></summary>
  
  Having discussed what makes two things 'the same', we now discuss whether one thing can be 'different' - in particular, whether the meaning of a message is inherent to that message, or also dependent on a mechanism interpreting it. 

  For this discussion, Hofstadter introduces three levels of messages:
  1. The frame message, which simply communicates that something _is_ a message
  2. The outer message, which communicates how the message should be decoded
  3. The inner message, which is the 'intended' message

  In 'normal' messages, only the inner message is explicitly recognized, but when a message is more abstract or encoded, we start to need more help from the other levels. For example, to turn DNA into a creature, or decode a message from aliens, we will a) first need to understand whether (and where) there even is a message, b) then need to understand how to interpret this message, and c) finally need to actually understand the intended meaning. This is made even more difficult if the encoded message doesn't have the same structure as the meaning. For example, a creature might have a certain fingerprint without having a 'fingerprint-gene'. 

  Reading tip: you may also compare this to the distinction between formal systems and their interpretation in chapter 2 

  The main question Hofstadter asks in this chapter is "is the meaning of a message located entirely in the message?". For something like a book, it seems logical that the meaning remains the same even if it is read by someone else or somewhere else. But is the same true for a strand of DNA that might develop completely differently when placed in a different chemical environment? Hofstadter's answer: some forms of messages are so inherent/universal that all forms of intelligence will be able to understand them, and hence (only) these messages can be said to have an 'intrinsic' meaning. 

  Reading tip: the idea of having to communicate with aliens is a good aid for understanding the three levels of a message. Try thinking how you would do such communication: what would make clear that there _is_ a message, how would the aliens learn how to _decode_ it, and how would they then _understand_ it? 
</details>

<details>
  <summary><i>Chapter VII: The Propositional Calculus</i></summary>
  
  This chapter introduces one more (and perhaps the most famous) formal system: that of the propositional logic, a formal language that attempts to formalize human reasoning. It does so by formalizing the concepts of 'and', 'or', 'not' and 'if...then...'.

  Reading tip: if you've taken a course in formal logic, you've probably seen this system before, although note that Hofstadter very consciously introduces very little rules (like only one De Morgan rule).

  Using this system, we can create derivations for statements like 'if x is true, and we know that 'if x then y', then y must be true'. Furthermore, the system does so purely typographically, so we can do so regardless of the meaning/content of atoms like 'x' and 'y'. Another nice property of the system is that is has a decision procedure (although it is not covered in the book).

  Of course, it'd be nice to make some formal claims _about_ the system as well. But doing so would be creating a metatheory, about which we could create a metametatheory, and so on. Doing so can get us in trouble - although propositional logic is a solid, simple base for more complicated systems.

  Finally, a note about contradiction. In propositional logic, if we even derive a contradiction ('p and not p'), we can conclude everything we want to. This idea of 'everything follows from a contradiction' is one of the reasons why we might want to avoid contradictions at all costs in our formal systems, even though we as humans can handle such contradictions in our normal thinking.

  Reading tip: this chapter might make clearer why concepts created in earlier chapters, such as concepts like decision procedures or the discussion from the Two-Part Intervention, are such a big deal for logicians.
</details>

<details>
  <summary><i>Chapter VIII: Typographical Number Theory</i></summary>
  
  One more formal system, then? This chapter introduces the typographical number theory (TNT), a system intended to express all facts about number theory. It does so by formalizing concepts like "is equal to", "larger than" and "plus", as well as "for all" and "there exists". TNT essentially encodes the five Peano-axioms, five "semi-formal" statements about the natural numbers (0, 1, 2, 3,...) made by Giuseppe Peano. 

  Reading tip: for those familiar with formal logic: this is essentially first-order logic with equality and a successor-predicate. 

  We encounter some problems, though. First of all, we notice that we can derive all statements of a certain form (0+0=0, 0+1=1, 0+2=2,...) but not the general form (0+n=n). Because of this, we conclude that TNT is a so-called omega-inconsistent system (although note that this is less bad than 'normal' inconsistency. In fact, we could introduce "not 0+n=n" without introducing inconsistency). 

  The bigger problem: how can we prove the (general) consistency of TNT? One way we might try this is by proving it using a less "strong" system, which doesn't use facts from number theory. This is something that Hilbert's programme tried. To no avail: Gödel proved that any system strong enough to prove the consistency of TNT, must be just as strong as TNT itself. 

  Reading tip: the chapter itself contains additional exercises, which you can use to understand TNT. Alternately, try understanding the Peano axioms first and then use those to understand TNT. 
</details>

<details>
  <summary><i>Chapter IX: Mumon and Gödel</i></summary>
  
  We start with Zen Buddhism. One of its basic tenets is that there is no way to characterize Zen, since no words can capture truth. Yet studying Zen is still worth our time, and this can be done through Koans. These are supposed to be triggers, which create confusion and encourage the listener to 'thread outside words/logic' - potentially leading to enlightenment (although enlightenment isn't the final goal of Zen).

  Hofstadter compares this with formal systems: words/formal systems can express some truths, but not all. This can also be compared to Escher and Bach: we can express some characteristic of their works using words/logic, but inevitably miss out on describing other parts as well. The rest of this chapter will explain why this is the case.

  Disclaimer: my understanding of Zen Buddhism is considerably smaller than that of logic, so take this reading guide with a grain of salt.

  Back to Hofstadter's MU. While we couldn't solve the MU-puzzle in the MIU-system directly, we can do so using number theory. Using a process using Gödel-numbering, we can express theorems of the MIU-system as numbers, and reason about them using number theory.

  In fact, we can do so for every formal system! Using Gödel-numbering, we can translate a formal system into typographical operations on numbers, and from there into mathematical operations on numbers. In short: a number can be both a symbol and a theorem, (i.e., there is an isomorphism between these).

  But if we can do so for every formal system, and we expressed number theory as the formal system TNT, then it follows that we can also Gödel-number TNT itself. But if theorems about TNT express facts about numbers, _and_ theorems about TNT can be Gödel-numbered into numbers, then theorems of TNT can also express facts about theorems about TNT itself. It other words, TNT 'tries to swallow itself'.

  This is where the final trick comes into play: if this is the case, then we can create some string G that expresses "G is not a theorem of TNT". But if this is a theorem  of TNT, then it must also be false (and hence lead to inconsistency). And since we assumed TNT never lead to falsity, we must conclude that G is true - and hence, we have something that is true, but not a theorem of TNT.

  Reading tip: how we can create G is explained in future chapters. For now, assume that is does exist, and try to understand how this results in inconsistency or incompleteness for TNT.
</details>


## Part 2: EGB

<details>
  <summary><i>Chapter X: Levels of Description, and Computer Systems</i></summary>
  
  We've seen that TNT and Gödel's theorem can be understood on different levels, and the same is true for many things, such as fugues. Sometimes this causes confusion, but other times it's no problem at all - we have no problem distinguishing humans from the cells composing them. This can be made possible by considering "chunks" of the individual components, such as organs as groups of cells. 

  Reading tip: the comparison to human biology can be continued throughout this chapter, and might makes things intuitively understandable

  Hofstadter compares this to programming languages. At their core, these languages consist of machine languages - off/on-states, "ones and zeroes" - yet these can be "grouped together" in assembly language and even further in higher-level languages. Compilers and interpreters can "translate" from high to low-level languages. At some point, we might ignore the lower levels entirely, e.g., by "bootstrapping" compilers, writing them in a compiled language. 

  Reading tip: you can compare this to the chapter about push/pop: one command in a high-level language "pushes" multiple, low-level commands, after which we "pop" back to the high level. 

  Such high-level languages - whether they be programming languages or something else - make things easier, and give us more "room for thought": a high-level language suited for some purpose makes expressing things related to that purpose easier. Yet we also lose some determinism: we no longer know exactly what happens at the lower level (which might differ per situation, e.g., per type of computer), which might result in outputs we couldn't predict with certainty.
</details>


<details>
  <summary><i>Chapter XI: Brains and Thoughts</i></summary>
  
  Reading tip: did you see the isomorphism between the ant fugue in the book and the fugue discussed _in_ the ant fugue?

  In trying to create artificial intelligence, we learned a lot about how human intelligence works. This chapter describes how human intelligence, in particular vision, might work. In the brain, descriptions don't have to be pinned to specific objects - we say that thought is _intensional_. But how does that work?

  The smallest element of the brain, the metaphorical 'ant' from the ant fugue, is the neuron, cells which input and output signals to other neurons. Can these neurons be combined to well-defined modules? Some evidence suggests that we should be able to, but scientists haven't done so as of the writing this book. 

  Nevertheless, Hofstadter suggests calling such modules "symbols", because unlike neurons, they symbolize concepts in the real world. In other words, there is a (fine-grained) isomorphism between these symbols and the world. These symbols, Hofstadter suggests, would be about the "size" of  - i.e., would represent things like - a word or phrase.

  So then, would these symbols represent (general) classes or (specific) instances - e.g., would a symbol represent all newspapers, or the specific newspaper that I have today? Hofstadter suggests both can be true: specific examples can 'split off' from their prototypical class and become symbols representing classes in their own right. But considering this in detail shows many complications, and shows that considering the brain as a well-defined state of symbols and activations is too simplistic.

  So now we have two questions: (how) can neurons create symbols, and how can symbol activation be explained in terms of itself? If the latter is true, we could also create intelligence on non-human hardware, since in that case we could create such symbols without depending on neurons. 
</details>

<details>
  <summary><i>Chapter XII: Minds and Thoughts</i></summary>
  
  Having introduced the concept of symbols, this chapter uses it to ask more questions about the brain and thought. For example: can we find an isomorphism between different human brains, or perhaps even between humans and animals? Definitely not on a neural level, nor an exact isomorphism, but perhaps a partial isomorphism on symbol-level? But what would a partial isomorphism even be? For example, would we be looking for similar characteristics on a local level, or on a global level? 

  Hofstadter illustrates how this might work using multiple examples, like the poem Jabberwocky or the map of Aporia. Not all details can be directly translated, but the relationship of details amongst themselves appears to be roughly similar - the feeling of the poem remains the same, and navigating Aporia is still possible, even if small details may have changed in translation. We can see this similarity as a rough isomorphism

  Using this concept of rough isomorphisms, we can compare systems like brains, languages and programming languages. 

  This creates a new question: is there a way to "read" a system at a high-level (e.g., read thoughts based on the symbols active in a brain)? Hofstadter claims that this is possible, although we would need a language to describe these states in. For example, which thoughts are real - describe a persons actual beliefs - and which are just imagination. 

  One more question would be where the "self" is in this system of symbols. Hofstadter claims that the self would be a subsystem of the brain: a highly sophisticated group of systems able to interact both with each other and the rest of the brain. 
</details>

<details>
  <summary><i>Chapter XIII: BlooP and FlooP and GlooP</i></summary>
  
  We have seen that both formal systems (e.g., the self-reference in TNT) and human thought (the 'self' as part of the brain) involve self-reference. For formal-systems, this self-reference was only relevant for 'sufficiently strong' formal systems. But what does that mean? For formal systems, a sufficient condition for being 'sufficiently strong' is containing all primitive recursive truths. This chapter aims to illustrate this concept, as well as related terms.

  To do so, the chapter introduces a number of programming languages. The first one of these is BlooP. BlooP can do some things typical of programming language: adding, multiplying and comparing numbers, branch based on 'if-statements', and do things multiple times using 'loops'. Its defining characteristic, however, is that these loops can only 'loop' a pre-defined number of times. In other words, if you tell a BlooP-programme to 'do something n times', then the number n must be known before starting the loop!

  Reading tip: the book provides a syntax for the programming languages, but this really doesn't matter too much, so skip it if necessary - it's the expressiveness of the different languages that matters most

  It turns out the functions one can express in BlooP are precisely the primitive recursive functions. In other words, TNT, being sufficiently strong, has all positive BlooP tests as theorems, and all negative BlooP tests as nontheorems. But does that mean everything can be expressed as a BlooP test? It turns out that is not true. Using a method called the _diagonal method_, we can create functions which are well-defined, but also (provably) not valid BlooP programmes. 

  What if we drop the boundedness of loops that defines BlooP? This language might be stronger, but it has a downside: it might contain programmes that never stop. And, in fact, Turing showed that we can also never know whether a programme will stop or not (or at least, there is no general-purpose test that decides this for any programme). Should we move on to an even stronger language, then? The so-called _Church-Turing Thesis_ argues that such languages do not exist - and, at the very least, we know that no such language has ever been found.
</details>

<details>
  <summary><i>Chapter XIV: On Formally Undecidable Propositions of TNT and Related Systems</i></summary>
  
  Remember when chapter 9 claimed we could create some sentence 'G' that could lead to either inconsistency or incompleteness? This chapter discusses how it is that we can create such a sentence in TNT.

  The core of the idea is what Hofstadter calls 'arithmoquining'. This consists of taking a sentence in TNT - in particular, a sentence with one variable - and 'inserting' that sentence into itself using Gödel numbering. This results in the sentence 'talking about itself', the form of self-reference that results in Godel's theorems.

  Reading tip: once again, the dialogue preceding the chapter gives a good introduction on what the chapter discusses formally. The main difference to keep in mind is that G does while being a number, instead of text. 

  In fact, Gödel's second incompleteness theorem states that no consistent system can prove it's own consistency. This is also where omega-inconsistency comes back: we cannot prove G, which is a claim about all natural numbers, yet for each _individual_ number we _can_ prove this claim.

  So what do we do now? We could consider expanding TNT in two ways: either by accepting G as a new axiom (the 'standard' extension), or by accepting the negation of G as a new axiom. The latter seems contradictory - how can something be true for all individual natural numbers but not for all numbers? - yet we handle this by accepting 'supernatural numbers', numbers that exist 'beyond' infinity.
</details>

<details>
  <summary><i>Chapter XV: Jumping out of the System</i></summary>
  
  So why don't we just add G to our formal system? Well, it turns out that in this formal system - TNT+G - we can repeat the same trick to create new undecidable sentences G'. And so on, ad infinitum. The same happens if we add the negation of G, or even if we try to add the entire "pattern" of G at once. TNT can be said to suffer from an _essential incompleteness_, one that cannot be eliminated. 

  This fact was used by J.R. Lucas to argue that the human mind cannot be interpreted as a machine (i.e., that there are things that humans can do but computers cannot). For we as humans can use this mechanism to _always_ create a statement that is true, but cannot be generated (i.e., "understood") by computers. 

  But is this true? Hofstadter argues it is not. If there is no algorithm that can describe how to "Gödelize" any system, then at one point this procedure will become too complicated for humans as well. 
</details>

<details>
  <summary><i>Chapter XVI: Self-Ref and Self-Rep</i></summary>
  
  This chapter discusses when self-reference may occur, and how that relates to self-replicating systems. 

  What is self-rep? We'd usually agree that a photo isn't self-replicating because it can be placed in a photo-copier. In other words, we want the self-replicating behavior to be defined explicitly. But, Hofstadter mentions, explicitness is not an all-or-nothing behavior. For example, we can create a self-rep programme that produces a self-rep "under translation" - which has the same internal message but in a different outer message. In fact, doing so in BlooP or BlooP requires this, because the output can only be numbers.

  Reading tip: this relates to the computing concept of quines, <a href="https://en.wikipedia.org/wiki/Quine_(computing)">the Wikipedia page</a> of which is quite interesting: 

  Another question we may ask is "where is the self-rep?". For example, is it just a programme, or also the interpreter interpreting the programme? The rest of this chapter will mostly discuss self-reps where these things are intertwined. 

  The chapter then introduces _typogenetics_, a simplified (formal) system discussing molecular genetics. Hofstadter compares TNT to molecular biology through typogenetics. Like how TNT is required to be "sufficiently strong" to facilitate self-ref, molecular genetics needs to be "sufficiently supported" by enzymes and ribosomes (like mRNA and RNA-polymerase) to facilitate self-rep.  Similar to how in the Quine-sentence one part functioned as an instruction and the other as an argument, DNA can function as both a programme and an argument. 

  Reading tip: see also the end of chapter 6, which asked the same question about the location of meaning. Here, you could see self-ref/self-rep as (part of) the meaning

  If TNT corresponds to molecular biology, what does Gödel's theorem correspond to? Hofstadter shows that (in spirit) the molecular equivalent to Gödel's theorem means that preventing viruses from doing harm is impossible.  
</details>

<details>
  <summary><i>Chapter XVII: Church, Turing, Tarski, and Others</i></summary>
  
  This chapter discusses the following theorem and its philosophical consequences: _every aspect of reasoning can be considered a high-level description of a system that, at a lower level, can be seen as a formal system_. For example, the brain can be seen as a kind of 'mathematical' object - although not in any way that a mathematician would accept. 

  The chapter first discusses some important ideas from mathematics and theoretical computer science, starting with the Church-Turing Thesis. This thesis, which can perhaps better be called a conjecture, essentially claims that every (decision) problem can be solved using a recursive function (i.e., a FlooP programme). So any decision problem - theorem vs non-theorem, beauty vs non-beauty, good vs evil - can eventually be reduced to simple operations like summation, multiplication and equation. 

  Reading tip: it's important to keep in mind that the Church-Turing-thesis is at its core a mathematical claim, so it does not claim that good and evil are antithetical or that everything is either beautiful or not - only that if they are (by some method), then that method can also be done using mathematics

  Hofstadter illustrates this with two examples: the mathematics of Ramanujan and the abilities of 'idiots savants'. He then strengthens the Church-Turing Thesis by adding the consequence that the recursive function/FlooP-programme is in some way isomorphic to the original 'owner' of the decision procedure, e.g. a human. 

  This, in turn, can be seen as an argument for the possibility that the behavior of some possibly living component (like a cell, or a neuron) can be accurately simulated by a computer program - but only when given a sufficiently accurate description of its internal and external state. 

  Fundamental skeptics of AI still have a few possible counterarguments, however. For example: 'is a concept like "beauty" not an irrational concept that can never be grasped by 'rational' computers'? Here, Hofstadter counters by claiming that "rationality" of computers at a hardware level does not imply rationality at a higher level - for example, computers can be instructed to flawlessly follow some completely irrational instructions. Likewise, the neuronal expressions of concepts like beauty are completely rational. 

  Hofstadter gives another interpretation for the 'complexity' of such concepts: these concepts, which he calls 'semantic', cannot be tested within a predetermined amount of time (i.e., with a BlooP-programme), because it's meaning is not 'localized' but connected with other concepts (e.g., a song might be beautiful because of associations with the place we heard it) 

  Likewise, Lucas' claim that Gödel makes AI impossible can be countered that although AI might be a formal system on one level, they might not be on another, and so Gödel's theorems no longer apply. 
</details>

  <summary><i>Chapter XVIII: Artificial Intelligence: Retrospects</i></summary>
  
  This chapter gives a short historical summary of the field of AI up to the publication of the book in 1979.

  Reading tip: keep in mind that the way AI systems are approached has changed significantly since 1979, so while historically important this should not be taken as a reflection on modern AI.

  The chapter starts by introducing the most famous "do AI systems think?"-test: the _imitation game_ by Alan Turing, now better known as the _Turing Test_. In this test, the AI system must act in such a way that a human judge cannot distinguish their (text-based) responses from a human doing the same. The chapter also discusses a number of objections, perhaps the most important one being the claim that an AI system that does not 'think' may still be able to bluff their way to 'passing' using the human's gullibility.

  The chapter then gives a brief history of AI, from the perspectives of AI doing mathematics, chess, creating language, and composing music. A common observation, dubber Tesler's Theorem by Hofstadter, is that all these things seem indicative of intelligence only until AI reaches them, in which case they're not 'real thinking' anymore.

  Hofstadter also discusses how the 'products' of AI systems should be interpreted. For example, if a human creates an AI system that creates a song, should the human be considered the 'author' or only the 'meta-author'? And if an AI system writes a sentence, to what do these words refer?

  Reading tip: this last point is essentially (or at least close to) the so-called _symbol grounding problem_

  <summary><i>Chapter XIX: Artificial Intelligence: Prospects</i></summary>
  
  Here, Hofstadter links artificial intelligence to counterfactuals, hypothetical scenario's in which some (but not all) features 'slip'. Hofstadter uses these counterfactuals to describe how artificial intelligence might develop after 1979

  Reading tip: since we're now after 1979, this makes any description of "future" AI that didn't come true a counterfactual itself 🙃.

  To start, Hofstadter describes the terms _constant, parameter_, and _variable_. Although the name suggests that some of these are always constant, the truth of the matter is that even constants can change - but only when the parameters and variables change as well, while this is not true in the other direction.

  This is linked to the concept of _frames_ in the field of AI: 'contexts' in which some values are fixed and some values can be changed. Furthermore, frames might have 'default' values for some values. However, by 'zooming in' or 'zooming out' to subframes or superframes, what is or isn't a constant might be changed, and this can be used to solve problems. Hofstadter illustrates this using the concept of the _Bongard Problem_, a set of 'find-the-similarity' tasks.

  This is then linked to other concepts discussed earlier in the book: language, enzymes, and symbols, as well as the creation of sections of the book itself. Hofstadter discusses whether creativity or intelligence can be linked to how often such 'slipping' can (or must) happen. This is then linked back to AI, finishing with Hofstadter speculatively answering ten questions about AI.
</details>

<details>
  <summary><i>Chapter XX: Strange Loops, Or Tangled Hierarchies</i></summary>
  
  In the final chapter, Hofstadter links Strange Loops to consciousness and AI. Consciousness and reasoning, he claims, are possible due to a multi-layered system, where the top layers are adaptable, but the bottom layers aren't - self-adapting software and unchanging hardware. 

  On every level, there is some 'protected' other level that (initially) cannot be changed. Sometimes it eventually turns out that this protected level _can_ be changed - but in doing so, we always create some new 'protected' level. When the levels also interact amongst themselves, a Strange Loop might be created. 

  Reading tip: try thinking about what the 'protected' and 'unprotected' levels might be in the other hierarchies discussed in the book (DNA, anthills, programming languages, etc.)

  Of course, when we encounted Strange Loops, Gödel's theorem comes into play. Hofstadter warns against translating Gödel's theorem to other fields directly, but also indirectly relates it to many fields, like Zen, science and art. 

  Does Gödel's theorem result in any limitations on our understanding of our mind? Hofstadter claims not, nor does 'Gödel' necessarily prevent the creation of AI. Quite the opposite might be true: a Gödelian mechanism in DNA makes the transfer of intelligence possible, after all. 

  In analogy to Gödel's G, Hofstadter claims that concepts like consciousness and free will might depend on the interplay between different 'levels'. Understanding ourselves, then, would require understanding the Tangled Hierarchy of levels in our mind. Likewise, Hofstadter believes this understanding will be the core of AI.
